```python
bins = [0,1,2,3,4,5]

labels = [
    "0–1 Poor",
    "1–2 Below Avg",
    "2–3 Average",
    "3–4 Good",
    "4–5 Excellent"
]

df["Rating Range"] = pd.cut(
    data["Aggregate rating"],
    bins=bins,
    labels=labels
)
```
